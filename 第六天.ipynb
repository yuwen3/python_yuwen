### 基本爬蟲

```python
import requests

from bs4 import BeautifulSoup
```

```python
url ="https://www.marieclaire.com.tw/love-sex/astrology"
resp =requests.get(url) 
```

```python
resp.text
```

```python
print(resp.text)  #看會不會比較漂亮，結果:不會
```

```python
soup =BeautifulSoup(resp.text,'lxml') 
#lxml整頓網頁
```

```python
hot =soup.find_all('div',class_="hot")
```

```python
articles =hot[0].find_all('a')
```

```python
len(articles)
```

```python
item =articles[0]
```

```python
item.attrs
```

```python
for item in articles:
    if "星星教授安格斯" in item['title']:
        print(item['title'],item['href'])
        weekly_url=item['href']
        #href把每一次都記下來
        print()
```

```python
weekly_url
```

```python
web =requests.get(weekly_url)
```

```python
weekly = BeautifulSoup(web.text,'lxml')
```

```python
weekly
```

```python
article =weekly.find_all('div',class_='articleContent')
```

```python
article
```

```python
article =weekly.find_all('div',class_='articleContent')[0]
```

```python
article.find_all('h2')
```

```python
fortune ={}

for item in article.find_all('h2'):
    print(item.string)
    luck =item.next_sibling
    print(luck.string)
```

```python
fortune ={}

for item in article.find_all('h2'):
    print(item.string[:2])
    luck =item.next_sibling
    print(luck.string)
```

```python
fortune ={}

for item in article.find_all('h2'):
    sign =item.string[:2]
    luck =item.next_sibling.string   #string去
    fortune[sign] =luck
```
